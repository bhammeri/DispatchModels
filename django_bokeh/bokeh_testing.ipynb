```python
import bokeh.plotting as bplt
import bokeh as bk
import pandas as pd
```

```python
station_info = pd.read_csv('station_info.csv')
station_info
```

```python
from bokeh.models import ColumnDataSource, Plot, LinearAxis, Grid
from bokeh.models.glyphs import Line, VBar
from bokeh.io import curdoc, show
```

```python
source = ColumnDataSource(station_info)
```

```python
plot = Plot(
    title=None, plot_width=300, plot_height=300,
    min_border=0, toolbar_location=None)

glyph = Line(x="index", y="ridership", line_color="#f46d43", line_width=6, line_alpha=0.6)
plot.add_glyph(source, glyph)

glyph =  VBar(x="index", top="ridership", bottom=0,)
plot.add_glyph(source, glyph)

xaxis = LinearAxis()
plot.add_layout(xaxis, 'below')

yaxis = LinearAxis()
plot.add_layout(yaxis, 'left')

plot.add_layout(Grid(dimension=0, ticker=xaxis.ticker))
plot.add_layout(Grid(dimension=1, ticker=yaxis.ticker))

curdoc().add_root(plot)

show(plot)
```
